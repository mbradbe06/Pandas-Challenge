## Imports

```python
# Dependencies and Setup
import pandas as pd

# File to Load 
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
Total_Players = len(purchase_data["SN"].unique())
Total_Players_df = pd.DataFrame({"Total Players":[Total_Players]})
Total_Players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Basic Calculations
Unique_Items = len(purchase_data["Item ID"].unique())
Average_Price = round(purchase_data["Price"].mean(),2)
Total_Purchases = purchase_data["Purchase ID"].count()
Total_Rev = purchase_data["Price"].sum()

#Summary DataFrame
Summary_df = pd.DataFrame({"Number of Unique Items": [Unique_Items],
                           "Average Price": f"${Average_Price}",
                           "Number of Purchases": [Total_Purchases],
                           "Total Revenue": f"${Total_Rev}"})

Summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#remove duplicate SN entries to get unique Gender data
purchase_data_dedup = purchase_data.drop_duplicates(subset ="SN", keep = 'first')

Genders = purchase_data_dedup["Gender"].value_counts()

#Generate new DataFrame with Genders
Gender_df = pd.DataFrame(Genders)

#Convert to percentages
Percentage_Male = round(Gender_df.iloc[0,0]/Total_Players*100,2)
Percentage_Female = round(Gender_df.iloc[1,0]/Total_Players*100,2)
Percentage_Other = round(Gender_df.iloc[2,0]/Total_Players*100,2)

#add Percentages Column
Gender_df["Percentages"] = [f"{Percentage_Male}%",
                            f"{Percentage_Female}%",
                            f"{Percentage_Other}%"]
Gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Group by Gender
Gender_Group = purchase_data.groupby(["Gender"])

#Clac Purchase Count, Avg. Purchase Price, and Total Purchase Value by Gender
Gender_Purchases = Gender_Group["Purchase ID"].count()
Gender_Avg_Price = round(Gender_Group["Price"].mean(),2)
Gender_TotalValue = round(Gender_Group["Price"].sum(),2)

#Calculate Avg Total Purchase per Person by Gender
Gender_TotalValue_df = pd.DataFrame(Gender_TotalValue)
Male_AvgPerPerson = round(Gender_TotalValue_df.iloc[1,0]/Gender_df.iloc[0,0],2)
Female_AvgPerPerson = round(Gender_TotalValue_df.iloc[0,0]/Gender_df.iloc[1,0],2)
Other_AvgPerPerson = round(Gender_TotalValue_df.iloc[2,0]/Gender_df.iloc[2,0],2)

#Format new DataFrame with all the info and format to currency
Gender_Summary_df = pd.DataFrame({"Purchase Count": Gender_Purchases,
                                  "Average Purchase Price": Gender_Avg_Price,
                                  "Total Purchase Value": Gender_TotalValue,
                                  "Avg Total Purchase per Person":
                                      {"Female": f"${Female_AvgPerPerson}",
                                       "Male": f"${Male_AvgPerPerson}",
                                       "Other / Non-Disclosed": f"${Other_AvgPerPerson}" }})

Gender_Summary_df["Average Purchase Price"] = Gender_Summary_df["Average Purchase Price"].map("${:.2f}".format)
Gender_Summary_df["Total Purchase Value"] = Gender_Summary_df["Total Purchase Value"].map("${:.2f}".format)
Gender_Summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#remove duplicate SN entries to get unique Age data
purchase_data_dedup = purchase_data.drop_duplicates(subset ="SN", keep = 'first')

#Create bins for ages
bins = [0,9,14,19,24,29,34,39,100]
Age_Ranges = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Categorize existing players with age bins
purchase_data_dedup["Age Ranges"] = pd.cut(purchase_data_dedup["Age"],bins,labels=Age_Ranges)

#Create a count of players by age group and format into dataframe  
#'.value_counts()' returns bins out of order so use '.reindex' to reorganize
#Change column title to "Total Count" for clarity
Purchase_age_ranges = purchase_data_dedup["Age Ranges"].value_counts()
Purchase_age_ranges_df = pd.DataFrame(Purchase_age_ranges)
Purchase_age_reorder_df = Purchase_age_ranges_df.reindex(["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"])
Purchase_age_reordername_df = Purchase_age_reorder_df.rename(columns = {"Age Ranges":"Total Count"})

#Add Percentages to DataFrame and format properly
Percentages = round(Purchase_age_reordername_df["Total Count"] / Total_Players * 100,2)
Purchase_age_reordername_df["Percentage of Players"] = Percentages
Purchase_age_reordername_df["Percentage of Players"] = Purchase_age_reordername_df["Percentage of Players"].map("{:.2f}%".format)
Purchase_age_reordername_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#bin purchase_data by age
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"],bins,labels=Age_Ranges)

#Get Purchase Counts by Age Range and organize format
Purchase_counts = purchase_data["Age Ranges"].value_counts()
Purchase_counts_df = pd.DataFrame(Purchase_counts)
Purchase_ReOrgCounts_df = Purchase_counts_df.reindex(["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"])
Purchase_ReOrgCounts_df = Purchase_ReOrgCounts_df.rename(columns={"Age Ranges":"Purchase Counts"})

#Tally avg purch price by age range, total purchase value by age range, 
# and avg total purchases per person by age range
Age_Range_Groups = purchase_data.groupby("Age Ranges")
Age_Range_AvgPurch = Age_Range_Groups["Price"].mean()
Purchase_ReOrgCounts_df["Average Purchase Price"] = Age_Range_AvgPurch

Age_Range_TotPurch = Age_Range_Groups["Price"].sum()
Purchase_ReOrgCounts_df["Total Purchase Value"] = Age_Range_TotPurch

Purchase_ReOrgCounts_df["Avg Total Purchase Per Person"] = (Purchase_ReOrgCounts_df["Total Purchase Value"] / 
                                                            Purchase_age_reordername_df["Total Count"]) 
#currency format
Purchase_ReOrgCounts_df["Average Purchase Price"] = Purchase_ReOrgCounts_df["Average Purchase Price"].map("${:.2f}".format)
Purchase_ReOrgCounts_df["Total Purchase Value"] = Purchase_ReOrgCounts_df["Total Purchase Value"].map("${:.2f}".format)
Purchase_ReOrgCounts_df["Avg Total Purchase Per Person"] = Purchase_ReOrgCounts_df["Avg Total Purchase Per Person"].map("${:.2f}".format)

#name index
Purchase_ReOrgCounts_df.index.name = "Age Ranges"
Purchase_ReOrgCounts_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Sort by "SN" to determine player spending data
grouped_sn = purchase_data.groupby(["SN"])

#format new DataFrame with calculations via GroupBy object functions to put into summary table
grouped_sn_df = pd.DataFrame({"Purchase Count": grouped_sn["Price"].count(),
                              "Average Purchase Price": grouped_sn["Price"].mean(),
                              "Total Purchase Value": grouped_sn["Price"].sum()})

#Sort spenders in descending order to find top spenders and currency format where necessary
top_sorted_df = grouped_sn_df.sort_values("Total Purchase Value", ascending = False)
top_sorted_df["Average Purchase Price"] = top_sorted_df["Average Purchase Price"].map("${:.2f}".format)
top_sorted_df["Total Purchase Value"] = top_sorted_df["Total Purchase Value"].map("${:.2f}".format)

#preview of top spenders
top_sorted_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Call Item/Price columns and group by Item ID/Name
Item_df = purchase_data[["Item ID", "Item Name", "Price"]]
GroupBy_Items = Item_df.groupby(["Item ID","Item Name"])

#format new DataFrame with calculations via GroupBy object functions to put into summary table
GroupBy_Items_df = pd.DataFrame({"Purchase Count": GroupBy_Items["Price"].count(),
                              "Item Price": GroupBy_Items["Price"].mean(),
                              "Total Purchase Value": GroupBy_Items["Price"].sum()})

#Sort Items in descending order to find most popular and currency format where necessary
Sorted_Items_df = GroupBy_Items_df.sort_values("Purchase Count", ascending = False)
Sorted_Items_df["Item Price"] = Sorted_Items_df["Item Price"].map("${:.2f}".format)
Sorted_Items_df["Total Purchase Value"] = Sorted_Items_df["Total Purchase Value"].map("${:.2f}".format)

#preview
Sorted_Items_df.head()
```
